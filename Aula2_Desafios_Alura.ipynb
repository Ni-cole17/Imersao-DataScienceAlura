```python
import pandas as pd
dados = pd.read_csv('https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true',compression = 'zip')
dados
```

```python
mapa = {'droga': 'composto'}
dados.rename(columns=mapa, inplace=True)
dados.head()
```

```python
compostos = dados['composto'].value_counts().index[0:5]
compostos
```

##**Desafio 1: Ordenar o grÃ¡fico countplot e plotar os outros grÃ¡ficos com o countplot**

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
plt.figure(figsize=(10,8))
ax = sns.countplot(x = 'composto', data = dados.query('composto in @compostos'))
ax.set_title("Top 5 Compostos")
plt.show()
```

```python

plt.figure(figsize=(10,8))
ax = sns.countplot(x = 'composto', data = dados.query('composto in @compostos'),order = compostos)
ax.set_title("Top 5 Compostos")
plt.show()
```

**Count**: Representa a quantidade total da ocorrÃªncia dos dados no G-0, que corresponde tambÃ©m a quantidade total de experimentos, visto que o G-0 foi analisado em todos os experimentos.

**Mean**: Representa a mÃ©dia dos valores localizados na coluna, nesse caso, o G-0

**Std**: Representa o desvio padrÃ£o, o qual Ã© uma medida de dispersÃ£o que avalia a uniformidade dos dados, logo quanto maior um desvio padrÃ£o mais sabemos que nossos dados estÃ£o espalhados.

**Min**: Representa o valor mÃ­nimo encontrado entre todos os valores da respectiva coluna

**25%**: Representa o primeiro quartil

**50%**: Representa a mediana

**75%**: Representa o terceiro quartil

**max**: Representa o valor mÃ¡ximo encontrado entre todos os valores da respectiva coluna

##**Desafio 5: Refletir sobre a manipulaÃ§Ã£o dos tamanhos das vizualizaÃ§Ãµes**

Para fazer essa analise e reflexÃ£o sobre a manipulaÃ§Ã£o das vizualizaÃ§Ãµes, testei algumas configuraÃ§Ãµes de visualizaÃ§Ã£o comeÃ§ando pela padrÃ£o definida pelo seaborn.

```python
sns.set()
ax = sns.countplot(x = 'composto', data = dados.query('composto in @compostos'), order = compostos)
ax.set_title("Top 5 Compostos")
plt.show()
```

Nessa configuraÃ§Ã£o "Default" como organizei de forma a ficar do maior para o menor, Ã© visÃ­vel saber qual dos compostos sÃ£o utilizados em maior quantidade em relaÃ§Ã£o aos outros. PorÃ©m dependendo do modo de vizualizaÃ§Ã£o desse grÃ¡fico pode tornar esse processo ou mais difÃ­cil ou mais fÃ¡cil para saber qual estÃ¡ em maior quantidade, principalmente aqueles que tem quantidades prÃ³ximas, como os 3 Ãºltimos.

Nesse grÃ¡fico abaixo percebe-se que o grid do background parece se aproximar mais do topo das barras, de um modo geral. Assim fica mais dificil nÃ£o sÃ³ de constatar entre os 2 Ãºltimos quem seria o maior, caso nÃ£o estivesse organizado como citado acima, como tambÃ©m de supor valores para as barras, em face da aproximaÃ§Ã£o observada.

```python
sns.set()
plt.figure(figsize=(30,5))
ax = sns.countplot(x = 'composto', data = dados.query('composto in @compostos'), order = compostos)
plt.xlabel('Doses',fontdict = {'fontsize':15})
plt.ylabel('Quantidade',fontdict = {'fontsize':15})
ax.set_title("Dosagem",fontdict = {'fontsize':25})
plt.show()
```

Nessa outra configuraÃ§Ã£o, fica mais difÃ­cil ainda a vizualizaÃ§Ã£o, como constatado anteriormente, as 3 Ãºltimas barras ficam mais parecidas ainda em altura.

```python
sns.set()
plt.figure(figsize=(100,10))
ax = sns.countplot(x = 'composto', data = dados.query('composto in @compostos'), order = compostos)
ax.set_title("Top 5 Compostos")
plt.xlabel('Doses',fontdict = {'fontsize':35})
plt.ylabel('Quantidade',fontdict = {'fontsize':35})
ax.set_title("Dosagem",fontdict = {'fontsize':55})
plt.show()
```

JÃ¡ nessa outra configuraÃ§Ã£o, em que hÃ¡ um alongamento vertical, fica mais fÃ¡cil notar diferenÃ§as entre as barras de comprimento "semelhante", Ã© possÃ­vel observar tambÃ©m que a barra do meio nÃ£o toca a grid da contagem de 250 como parecia nos grÃ¡ficos acima. PorÃ©m torna mais difÃ­cil a vizualizaÃ§Ã£o de todos os dados, sem ter que necessariamente estar focando em uma das barras. AlÃ©m disso,vÃª-se um efeito contrÃ¡rio do que acontecia quando se alongava horizontalmente, nesse caso hÃ¡ um maior distanciamento da grid com o topo das colunas.

```python
sns.set()
plt.figure(figsize=(10,20))
ax = sns.countplot(x = 'composto', data = dados.query('composto in @compostos'),order = compostos)
plt.xlabel('Doses',fontdict = {'fontsize':15})
plt.ylabel('Quantidade',fontdict = {'fontsize':15})
ax.set_title("Dosagem",fontdict = {'fontsize':25})
plt.show()
```

Nesse Ãºltimo, nota-se exatamente o que foi constatado na notaÃ§Ã£o acima, em que as diferenÃ§as sÃ£o mais nÃ­tidas e o comprimento de cada barra mais definido.

```python
sns.set()
plt.figure(figsize=(10,50))
ax = sns.countplot(x = 'composto', data = dados.query('composto in @compostos'), order = compostos)
plt.xlabel('Doses',fontdict = {'fontsize':15})
plt.ylabel('Quantidade',fontdict = {'fontsize':15})
ax.set_title("Dosagem",fontdict = {'fontsize':25})
plt.show()
```
