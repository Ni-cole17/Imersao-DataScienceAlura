```python
import pandas as pd
dados = pd.read_csv('https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true',compression = 'zip')
dados
```

```python
mapa = {'droga': 'composto'}
dados.rename(columns=mapa, inplace=True)
dados.head()
```

```python
compostos = dados['composto'].value_counts().index[0:5]
compostos
```

##**Desafio 1: Ordenar o gráfico countplot e plotar os outros gráficos com o countplot**

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
plt.figure(figsize=(10,8))
ax = sns.countplot(x = 'composto', data = dados.query('composto in @compostos'))
ax.set_title("Top 5 Compostos")
plt.show()
```

```python

plt.figure(figsize=(10,8))
ax = sns.countplot(x = 'composto', data = dados.query('composto in @compostos'),order = compostos)
ax.set_title("Top 5 Compostos")
plt.show()
```

##**Desafio 4: Estudar sobre as estatísticas retornadas no .describe.**

As estatísticas retornadas no .describe são as seguintes:

```python
dados['g-0'].describe()
```

**Count**: Representa a quantidade total da ocorrência dos dados no G-0, que corresponde também a quantidade total de experimentos, visto que o G-0 foi analisado em todos os experimentos.

**Mean**: Representa a média dos valores localizados na coluna, nesse caso, o G-0

**Std**: Representa o desvio padrão, o qual é uma medida de dispersão que avalia a uniformidade dos dados, logo quanto maior um desvio padrão mais sabemos que nossos dados estão espalhados.

**Min**: Representa o valor mínimo encontrado entre todos os valores da respectiva coluna

**25%**: Representa o primeiro quartil

**50%**: Representa a mediana

**75%**: Representa o terceiro quartil

**max**: Representa o valor máximo encontrado entre todos os valores da respectiva coluna

##**Desafio 5: Refletir sobre a manipulação dos tamanhos das vizualizações**

Para fazer essa analise e reflexão sobre a manipulação das vizualizações, testei algumas configurações de visualização começando pela padrão definida pelo seaborn.

```python
sns.set()
ax = sns.countplot(x = 'composto', data = dados.query('composto in @compostos'), order = compostos)
ax.set_title("Top 5 Compostos")
plt.show()
```

Nessa configuração "Default" como organizei de forma a ficar do maior para o menor, é visível saber qual dos compostos são utilizados em maior quantidade em relação aos outros. Porém dependendo do modo de vizualização desse gráfico pode tornar esse processo ou mais difícil ou mais fácil para saber qual está em maior quantidade, principalmente aqueles que tem quantidades próximas, como os 3 últimos.

Nesse gráfico abaixo percebe-se que o grid do background parece se aproximar mais do topo das barras, de um modo geral. Assim fica mais dificil não só de constatar entre os 2 últimos quem seria o maior, caso não estivesse organizado como citado acima, como também de supor valores para as barras, em face da aproximação observada.

```python
sns.set()
plt.figure(figsize=(30,5))
ax = sns.countplot(x = 'composto', data = dados.query('composto in @compostos'), order = compostos)
plt.xlabel('Doses',fontdict = {'fontsize':15})
plt.ylabel('Quantidade',fontdict = {'fontsize':15})
ax.set_title("Dosagem",fontdict = {'fontsize':25})
plt.show()
```

Nessa outra configuração, fica mais difícil ainda a vizualização, como constatado anteriormente, as 3 últimas barras ficam mais parecidas ainda em altura.

```python
sns.set()
plt.figure(figsize=(100,10))
ax = sns.countplot(x = 'composto', data = dados.query('composto in @compostos'), order = compostos)
ax.set_title("Top 5 Compostos")
plt.xlabel('Doses',fontdict = {'fontsize':35})
plt.ylabel('Quantidade',fontdict = {'fontsize':35})
ax.set_title("Dosagem",fontdict = {'fontsize':55})
plt.show()
```

Já nessa outra configuração, em que há um alongamento vertical, fica mais fácil notar diferenças entre as barras de comprimento "semelhante", é possível observar também que a barra do meio não toca a grid da contagem de 250 como parecia nos gráficos acima. Porém torna mais difícil a vizualização de todos os dados, sem ter que necessariamente estar focando em uma das barras. Além disso,vê-se um efeito contrário do que acontecia quando se alongava horizontalmente, nesse caso há um maior distanciamento da grid com o topo das colunas.

```python
sns.set()
plt.figure(figsize=(10,20))
ax = sns.countplot(x = 'composto', data = dados.query('composto in @compostos'),order = compostos)
plt.xlabel('Doses',fontdict = {'fontsize':15})
plt.ylabel('Quantidade',fontdict = {'fontsize':15})
ax.set_title("Dosagem",fontdict = {'fontsize':25})
plt.show()
```

Nesse último, nota-se exatamente o que foi constatado na notação acima, em que as diferenças são mais nítidas e o comprimento de cada barra mais definido.

```python
sns.set()
plt.figure(figsize=(10,50))
ax = sns.countplot(x = 'composto', data = dados.query('composto in @compostos'), order = compostos)
plt.xlabel('Doses',fontdict = {'fontsize':15})
plt.ylabel('Quantidade',fontdict = {'fontsize':15})
ax.set_title("Dosagem",fontdict = {'fontsize':25})
plt.show()
```
